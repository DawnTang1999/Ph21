```python
import urllib2
import urllib
import matplotlib.pyplot as plt
from astropy.io.votable import parse
```

```python
# Part 1 access webpage
url = "http://nesssi.cacr.caltech.edu/DataRelease/upload/result_web_file21Over.csv"
values = {}
req = urllib2.Request(url)
response = urllib2.urlopen(req)
the_page = response.read()
#print(the_page)
```

```python
# snip out the wanted parts (time MJD and magnitude mag)
lines = the_page.split('\n')
lines.pop(0)
time = []
mag = []
#print(lines[0].split(','))
for i in range(len(lines)):
    line = lines[i].split(',')
    if len(line) == 7:
        time.append(float(line[5]))
        mag.append(float(line[1]))
#print(time)
#print(mag)
# plot data
plt.plot(time, mag, 'o')
plt.xlabel("Time (MJD)")
plt.ylabel("Magnitude")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Part 2
# Get data from website
votable = parse("http://nesssi.cacr.caltech.edu/DataRelease/upload/result_web_fileY0xtZz.vot", pedantic = False)
table = votable.get_first_table()
data = table.array
#print data
# Create arrays for time and magnitude data
mag_fake_array = table.array['Mag']
time_fake_array = table.array['ObsTime']
mag = []
time = []
for i in range(len(mag_fake_array)):
    mag.append(mag_fake_array[i][0])
    time.append(time_fake_array[i][0])
#print(mag)
#print(time)
# Plot data
plt.plot(time, mag, 'o')
plt.xlabel("Time (MJD)")
plt.ylabel("Magnitude")
plt.gca().invert_yaxis()
plt.show()
```
